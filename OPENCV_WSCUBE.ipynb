```python
# for subtraction of images 
# now we will add both the images for that both images must have same size 
img1=cv2.imread("GK_image1.jpg")
img2=cv2.imread("GK_image3.jpg")
img1_resize1=cv2.resize(img1,(400,500))
img2_resize2=cv2.resize(img2,(400,500))
added_img=cv2.subtract(img1_resize1,img2_resize2)
cv2.imshow("Gulshan_image1",img1_resize1)
cv2.imshow("Gulshan_image3",img2_resize2)
cv2.imshow("GULSHAN_ADDED_IMAGE",added_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## BITWISE OPERATION ON IMAGES(BINARY)

```python
# this is how we can generate binary images which contain only either black or white from colourful images
# Read an image in grayscale
img1 = cv2.imread('GK_image4.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('GK_image3.jpg', cv2.IMREAD_GRAYSCALE)
img1_resize=cv2.resize(img1,(400,500))
img2_resize=cv2.resize(img2,(400,500))
# Apply binary thresholding
_, binary_img1 = cv2.threshold(img1_resize, 127, 255, cv2.THRESH_BINARY)
_, binary_img2 = cv2.threshold(img2_resize, 127, 255, cv2.THRESH_BINARY)
# cv2.imshow('Original1', img1_resize)
# cv2.imshow('Original2', img2_resize)
# cv2.imshow('Binary1', binary_img1)
# cv2.imshow('Binary2', binary_img2)
h1=np.hstack((img1_resize,binary_img1))
h2=np.hstack((img2_resize,binary_img2))
cv2.imshow("Gulshan_img1",h1)
cv2.imshow("Gulshan_img3",h2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# now bitwise operation on images
after_bitwise_and=cv2.bitwise_and(binary_img1,binary_img2)
h=np.hstack((binary_img1,binary_img2,after_bitwise_and))
cv2.imshow("Bitwise_and_operation",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#bitwise_or
after_bitwise_or=cv2.bitwise_or(binary_img1,binary_img2)
h=np.hstack((binary_img1,binary_img2,after_bitwise_and))
cv2.imshow("Bitwise_and_operation",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#bitwise_xor
after_bitwise_and=cv2.bitwise_xor(binary_img1,binary_img2)
h=np.hstack((binary_img1,binary_img2,after_bitwise_and))
cv2.imshow("Bitwise_and_operation",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#botwise_not
after_bitwise_and=cv2.bitwise_not(binary_img1)
h=np.hstack((binary_img1,after_bitwise_and))
cv2.imshow("Bitwise_and_operation",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## IMAGE SCALING, ROTATING USING OPENCV

```python
img=cv2.imread("RAK_image1.jpg")
img_resize=cv2.resize(img,(400,400))
#if both w&h are equal then only we can use (w,h) or(h,w) in cv2.warpAffine() function
# and if w&h are different while resizing the image then have to use (h,w) in cv2.warpAffine() function 
#because then only matrix multiplication is possible

#to get center use
width=img_resize.shape[0]
height=img_resize.shape[1]
new_image_after_scaling=cv2.getRotationMatrix2D((width/2,height/2),angle=90,scale=0.8)
#scale=1 then same if scale<1 then compresses image and if scale>1 then enlarges the image
final_image=cv2.warpAffine(img_resize,new_image_after_scaling,(width,height),borderValue=(0,0,255))
h=np.hstack((img_resize,final_image))
cv2.imshow("ROTATION_IMAGE",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#drawing Rectangle using OpenCV
drawing = False
ix, iy = -1, -1
img = np.ones((512, 512, 3), np.uint8)*255

def draw_rectangle(event, x, y, flags, param):
    global ix, iy, drawing

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            temp = img.copy()
            cv2.rectangle(temp, (ix, iy), (x, y), (0, 0, 255), 3)
            cv2.imshow("Draw_Rect", temp)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), 3)

cv2.namedWindow("Draw_Rect")
cv2.setMouseCallback("Draw_Rect", draw_rectangle)

while True:
    cv2.imshow("Draw_Rect", img)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# ============================================================================
# FIXED AND IMPROVED VERSION
# ============================================================================

# Read image (add proper file extension)
img = cv2.imread("contour")  # or .png, .jpeg

# Check if image was loaded successfully
if img is None:
    print("Error: Could not read image. Check the file path.")
    exit()

# Resize image
img = cv2.resize(img, (400, 500))

# Convert to grayscale
gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply threshold
_, thr = cv2.threshold(gry, 233, 255, cv2.THRESH_BINARY)

# Find contours
cnt, hir = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Create a copy to draw on (don't modify original)
result = img.copy()

# Store areas
ar = []

print(f"Number of contours found: {len(cnt)}")

for i, c in enumerate(cnt):
    # Calculate area first
    a = cv2.contourArea(c)
    ar.append(a)
    
    # Skip very small contours (noise)
    if a < 100:  # Adjust threshold as needed
        continue
    
    # Calculate moments
    m = cv2.moments(c)
    
    # Check if m00 is not zero to avoid division by zero
    if m["m00"] != 0:
        x = int(m["m10"] / m["m00"])
        y = int(m["m01"] / m["m00"])
        
        # Draw centroid
        cv2.circle(result, (x, y), 6, (255, 0, 0), -1)
        
        # Add label
        cv2.putText(result, f"{i}", (x + 10, y), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    
    # Draw contours
    cv2.drawContours(result, [c], -1, (0, 0, 255), 2)
    
    # Contour approximation
    ep = 0.1 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, ep, True)
    
    # Convex hull (use original contour, not approximation)
    hull = cv2.convexHull(c)
    
    # Draw convex hull
    cv2.drawContours(result, [hull], -1, (0, 255, 255), 2)
    
    # Bounding rectangle (FIXED: renamed variable to avoid conflict)
    x_rect, y_rect, w_rect, h_rect = cv2.boundingRect(hull)
    cv2.rectangle(result, (x_rect, y_rect), 
                 (x_rect + w_rect, y_rect + h_rect), (0, 255, 0), 2)
    
    # Print contour information
    print(f"\nContour {i}:")
    print(f"  Area: {a:.2f}")
    print(f"  Perimeter: {cv2.arcLength(c, True):.2f}")
    print(f"  Vertices (original): {len(c)}")
    print(f"  Vertices (approximated): {len(approx)}")
    print(f"  Bounding Box: ({x_rect}, {y_rect}, {w_rect}, {h_rect})")

# Display results
cv2.imshow("Original", img)
cv2.imshow("Result with Contours", result)
cv2.imshow("Binary Threshold", thr)

print(f"\nTotal contours: {len(cnt)}")
print(f"Areas: {ar[:10]}")  # Show first 10 areas

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python

# ============================================================================
# ALTERNATIVE: BETTER ORGANIZED VERSION
# ============================================================================

def analyze_contours(image_path, threshold_value=233, min_area=100):
    """
    Analyze contours in an image with proper error handling
    
    Args:
        image_path: Path to the image file
        threshold_value: Threshold value for binary conversion
        min_area: Minimum contour area to consider (filters noise)
    """
    # Read and validate image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Cannot read image from {image_path}")
        return None
    
    # Resize
    img = cv2.resize(img, (400, 500))
    original = img.copy()
    
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply threshold
    _, binary = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY)
    
    # Find contours
    contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, 
                                           cv2.CHAIN_APPROX_SIMPLE)
    
    print(f"Total contours found: {len(contours)}")
    
    # Process each contour
    areas = []
    valid_contours = 0
    
    for idx, contour in enumerate(contours):
        # Calculate area
        area = cv2.contourArea(contour)
        areas.append(area)
        
        # Skip small contours
        if area < min_area:
            continue
        
        valid_contours += 1
        
        # Moments and centroid
        moments = cv2.moments(contour)
        if moments["m00"] != 0:
            cx = int(moments["m10"] / moments["m00"])
            cy = int(moments["m01"] / moments["m00"])
            
            # Draw centroid
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), -1)
            cv2.putText(img, str(idx), (cx + 10, cy), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
        
        # Draw contour
        cv2.drawContours(img, [contour], -1, (0, 0, 255), 2)
        
        # Contour approximation
        perimeter = cv2.arcLength(contour, True)
        epsilon = 0.1 * perimeter
        approx = cv2.approxPolyDP(contour, epsilon, True)
        
        # Convex hull
        hull = cv2.convexHull(contour)
        cv2.drawContours(img, [hull], -1, (0, 255, 255), 1)
        
        # Bounding rectangle
        x, y, w, h = cv2.boundingRect(hull)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        # Check if convex
        is_convex = cv2.isContourConvex(contour)
        
        print(f"\nContour {idx}:")
        print(f"  Area: {area:.2f}")
        print(f"  Perimeter: {perimeter:.2f}")
        print(f"  Approx vertices: {len(approx)}")
        print(f"  Is convex: {is_convex}")
    
    print(f"\nValid contours (area > {min_area}): {valid_contours}")
    
    # Display
    cv2.imshow("Original", original)
    cv2.imshow("Processed", img)
    cv2.imshow("Binary", binary)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return contours, areas


# ============================================================================
# USAGE
# ============================================================================

# Method 1: Use the fixed inline code (lines 7-85)

# Method 2: Use the organized function
# analyze_contours("contour.jpg", threshold_value=233, min_area=100)
```

## IMAGE BACKGROUND REMOVAL HISTOGRAM

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ============================================================================
# GREEN SCREEN BACKGROUND REMOVAL WITH HISTOGRAM ANALYSIS
# ============================================================================

# Read the uploaded image
# Replace 'image.jpg' with your actual filename
img = cv2.imread("green_screen.png")

if img is None:
    print("Error: Could not load image. Please check the filename.")
    exit()

print(f"Image loaded successfully! Shape: {img.shape}")

# ============================================================================
# 1. ANALYZE HISTOGRAM
# ============================================================================

def analyze_image_histogram(img):
    """Analyze and display color histograms"""
    
    # Calculate histograms for BGR channels
    colors = ('b', 'g', 'r')
    color_names = ('Blue', 'Green', 'Red')
    
    plt.figure(figsize=(15, 4))
    
    for i, (color, name) in enumerate(zip(colors, color_names)):
        hist = cv2.calcHist([img], [i], None, [256], [0, 256])
        
        plt.subplot(1, 3, i + 1)
        plt.plot(hist, color=color, linewidth=2)
        plt.title(f'{name} Channel Histogram')
        plt.xlabel('Pixel Intensity')
        plt.ylabel('Frequency')
        plt.xlim([0, 256])
        plt.grid(True, alpha=0.3)
        
        # Find peak
        peak_val = np.argmax(hist)
        plt.axvline(peak_val, color='red', linestyle='--', 
                   label=f'Peak: {peak_val}')
        plt.legend()
    
    plt.tight_layout()
    plt.savefig('histogram_analysis.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("\nHistogram Analysis:")
    print(f"Green channel has highest values - confirming green screen")
    print("Histogram saved as 'histogram_analysis.png'")


# ============================================================================
# 2. METHOD 1: HSV COLOR RANGE (BEST FOR GREEN SCREEN)
# ============================================================================

def remove_green_background_hsv(img):
    """Remove green background using HSV color space"""
    
    # Convert to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Define range for green color
    # Hue for green is around 60 (in OpenCV: 0-179)
    lower_green = np.array([35, 40, 40])
    upper_green = np.array([85, 255, 255])
    
    # Create mask for green pixels
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    
    # Invert mask (we want to keep non-green pixels)
    mask_inv = cv2.bitwise_not(mask_green)
    
    # Refine mask with morphological operations
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask_inv = cv2.morphologyEx(mask_inv, cv2.MORPH_CLOSE, kernel, iterations=3)
    mask_inv = cv2.morphologyEx(mask_inv, cv2.MORPH_OPEN, kernel, iterations=2)
    
    # Apply mask to original image
    result = cv2.bitwise_and(img, img, mask=mask_inv)
    
    # Optional: Create transparent background (RGBA)
    # Add alpha channel
    b, g, r = cv2.split(img)
    rgba = cv2.merge([b, g, r, mask_inv])
    
    return result, mask_inv, rgba


# ============================================================================
# 3. METHOD 2: HISTOGRAM BACKPROJECTION
# ============================================================================

def remove_bg_histogram_backprojection(img):
    """Remove background using histogram backprojection"""
    
    # Convert to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Define ROI for foreground (center region with people)
    h, w = img.shape[:2]
    roi = img[int(h*0.4):int(h*0.95), int(w*0.15):int(w*0.7)]
    roi_hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
    
    # Calculate histogram of ROI (foreground)
    roi_hist = cv2.calcHist([roi_hsv], [0, 1], None, [180, 256], 
                           [0, 180, 0, 256])
    
    # Normalize histogram
    cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)
    
    # Calculate back projection
    dst = cv2.calcBackProject([hsv], [0, 1], roi_hist, 
                             [0, 180, 0, 256], 1)
    
    # Apply threshold
    _, mask = cv2.threshold(dst, 30, 255, cv2.THRESH_BINARY)
    
    # Clean up mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=3)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=2)
    
    # Apply mask
    result = cv2.bitwise_and(img, img, mask=mask)
    
    return result, mask, dst


# ============================================================================
# 4. METHOD 3: GRABCUT ALGORITHM
# ============================================================================

def remove_bg_grabcut(img):
    """Advanced background removal using GrabCut"""
    
    # Create mask
    mask = np.zeros(img.shape[:2], np.uint8)
    
    # Background and foreground models
    bgd_model = np.zeros((1, 65), np.float64)
    fgd_model = np.zeros((1, 65), np.float64)
    
    # Define rectangle around foreground
    h, w = img.shape[:2]
    rect = (int(w*0.1), int(h*0.35), int(w*0.65), int(h*0.6))
    
    # Apply GrabCut
    cv2.grabCut(img, mask, rect, bgd_model, fgd_model, 
                5, cv2.GC_INIT_WITH_RECT)
    
    # Modify mask
    mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
    
    # Apply mask
    result = img * mask2[:, :, np.newaxis]
    
    return result, mask2 * 255


# ============================================================================
# 5. METHOD 4: ADVANCED GREEN SCREEN REMOVAL
# ============================================================================

def advanced_green_removal(img):
    """Advanced green screen removal with edge refinement"""
    
    # Convert to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Multiple green ranges to catch all shades
    lower_green1 = np.array([35, 40, 40])
    upper_green1 = np.array([85, 255, 255])
    
    # Create mask
    mask1 = cv2.inRange(hsv, lower_green1, upper_green1)
    
    # Also check in RGB for pure green
    lower_green_rgb = np.array([0, 100, 0])
    upper_green_rgb = np.array([100, 255, 100])
    mask2 = cv2.inRange(img, lower_green_rgb, upper_green_rgb)
    
    # Combine masks
    mask = cv2.bitwise_or(mask1, mask2)
    
    # Invert mask
    mask_inv = cv2.bitwise_not(mask)
    
    # Edge detection for refinement
    edges = cv2.Canny(img, 50, 150)
    
    # Dilate edges slightly
    kernel = np.ones((3, 3), np.uint8)
    edges = cv2.dilate(edges, kernel, iterations=1)
    
    # Use edges to refine mask
    mask_refined = cv2.bitwise_and(mask_inv, cv2.bitwise_not(edges))
    
    # Morphological operations
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask_refined = cv2.morphologyEx(mask_refined, cv2.MORPH_CLOSE, 
                                    kernel, iterations=3)
    mask_refined = cv2.morphologyEx(mask_refined, cv2.MORPH_OPEN, 
                                    kernel, iterations=2)
    
    # Smooth edges with Gaussian blur
    mask_refined = cv2.GaussianBlur(mask_refined, (5, 5), 0)
    _, mask_refined = cv2.threshold(mask_refined, 127, 255, cv2.THRESH_BINARY)
    
    # Apply mask
    result = cv2.bitwise_and(img, img, mask=mask_refined)
    
    # Create with white background
    white_bg = np.ones_like(img) * 255
    result_white = np.where(result == 0, white_bg, result)
    
    # Create with transparent background
    b, g, r = cv2.split(img)
    rgba = cv2.merge([b, g, r, mask_refined])
    
    return result, result_white, mask_refined, rgba


# ============================================================================
# EXECUTE ALL METHODS
# ============================================================================

print("\n" + "="*70)
print("GREEN SCREEN BACKGROUND REMOVAL - PROCESSING")
print("="*70)

# Analyze histogram
print("\n1. Analyzing histogram...")
analyze_image_histogram(img)

# Method 1: HSV
print("\n2. Applying HSV color range method...")
result1, mask1, rgba1 = remove_green_background_hsv(img)

# Method 2: Histogram Backprojection
print("\n3. Applying histogram backprojection...")
result2, mask2, backproj = remove_bg_histogram_backprojection(img)

# Method 3: GrabCut
print("\n4. Applying GrabCut algorithm...")
result3, mask3 = remove_bg_grabcut(img.copy())

# Method 4: Advanced
print("\n5. Applying advanced green removal...")
result4, result4_white, mask4, rgba4 = advanced_green_removal(img)

# ============================================================================
# DISPLAY RESULTS
# ============================================================================

print("\n6. Displaying results...")

plt.figure(figsize=(20, 12))

# Row 1: Original and Methods
plt.subplot(3, 5, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize=12, fontweight='bold')
plt.axis('off')

plt.subplot(3, 5, 2)
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_BGR2RGB))
plt.title('Method 1: HSV Range', fontsize=12, fontweight='bold')
plt.axis('off')

plt.subplot(3, 5, 3)
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.title('Method 2: Backprojection', fontsize=12, fontweight='bold')
plt.axis('off')

plt.subplot(3, 5, 4)
plt.imshow(cv2.cvtColor(result3, cv2.COLOR_BGR2RGB))
plt.title('Method 3: GrabCut', fontsize=12, fontweight='bold')
plt.axis('off')

plt.subplot(3, 5, 5)
plt.imshow(cv2.cvtColor(result4, cv2.COLOR_BGR2RGB))
plt.title('Method 4: Advanced (BEST)', fontsize=12, fontweight='bold', color='green')
plt.axis('off')

# Row 2: Masks
plt.subplot(3, 5, 6)
plt.imshow(mask1, cmap='gray')
plt.title('Mask 1', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 7)
plt.imshow(mask2, cmap='gray')
plt.title('Mask 2', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 8)
plt.imshow(mask3, cmap='gray')
plt.title('Mask 3', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 9)
plt.imshow(mask4, cmap='gray')
plt.title('Mask 4', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 10)
plt.imshow(backproj, cmap='hot')
plt.title('Back Projection', fontsize=10)
plt.axis('off')

# Row 3: Final results with different backgrounds
plt.subplot(3, 5, 11)
plt.imshow(cv2.cvtColor(result4_white, cv2.COLOR_BGR2RGB))
plt.title('White Background', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 12)
# Create black background version
black_bg = np.zeros_like(img, dtype=np.uint8)
mask_3ch = cv2.cvtColor(mask4, cv2.COLOR_GRAY2BGR)
result_black = np.where(mask_3ch == 255, result4, black_bg).astype(np.uint8)
plt.imshow(cv2.cvtColor(result_black, cv2.COLOR_BGR2RGB))
plt.title('Black Background', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 13)
# Create blue background version
blue_bg = np.ones_like(img, dtype=np.uint8)
blue_bg[:] = [255, 200, 100]  # Light blue in BGR
result_blue = np.where(mask_3ch == 255, result4, blue_bg).astype(np.uint8)
plt.imshow(cv2.cvtColor(result_blue, cv2.COLOR_BGR2RGB))
plt.title('Blue Background', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 14)
# Show RGBA (transparency)
plt.imshow(cv2.cvtColor(rgba4, cv2.COLOR_BGRA2RGBA))
plt.title('With Transparency', fontsize=10)
plt.axis('off')

plt.subplot(3, 5, 15)
# Show edge detection
edges = cv2.Canny(img, 50, 150)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection', fontsize=10)
plt.axis('off')

plt.tight_layout()
plt.savefig('background_removal_results.png', dpi=150, bbox_inches='tight')
plt.show()

# ============================================================================
# SAVE RESULTS
# ============================================================================

print("\n7. Saving results...")

cv2.imwrite('result_hsv.png', result1)
cv2.imwrite('result_backprojection.png', result2)
cv2.imwrite('result_grabcut.png', result3)
cv2.imwrite('result_advanced.png', result4)
cv2.imwrite('result_white_bg.png', result4_white)
cv2.imwrite('result_transparent.png', rgba4)
cv2.imwrite('mask_final.png', mask4)

print("\n" + "="*70)
print("PROCESSING COMPLETE!")
print("="*70)
print("\nSaved files:")
print("  - background_removal_results.png (comparison)")
print("  - histogram_analysis.png")
print("  - result_hsv.png")
print("  - result_backprojection.png")
print("  - result_grabcut.png")
print("  - result_advanced.png (BEST)")
print("  - result_white_bg.png")
print("  - result_transparent.png (PNG with alpha)")
print("  - mask_final.png")
print("\nâœ… RECOMMENDED: Use 'result_advanced.png' or 'result_transparent.png'")
print("="*70)
```

## Hough Line Transformation

```python
#lengthy method
img=cv2.imread("Sudoko.png")
img_resize=cv2.resize(img,(500,500))
gry=cv2.cvtColor(img_resize,cv2.COLOR_BGR2GRAY)
#edgedetection
edg=cv2.Canny(gry,20,250)
#thetha for HoughLines
#make hough line
lines=cv2.HoughLines(edg,1,np.pi/180,200)
for r,theta in lines[10]:
    a=np.cos(theta) #acos(Theta)+bsin(theta)
    b=np.sin(theta)

    x0=a*r
    y0=b*r

    x1=int(x0+1000*(-b))
    y1=int(y0+1000*(a))
    x2=int(x0-1000*(-b))
    y2=int(y0-1000*(a))

    cv2.line(img_resize,(x1,y1),(x2,y2),(0,255,0),2)
cv2.imshow("Original_image",img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
